```python
import pandas as pd

dane = pd.read_csv('../data/ab_data.csv')
country = pd.read_csv('../data/countries.csv')
```

```python
dane
```

```python
country
```

```python
fulldata = dane.merge(country, on = 'user_id', how='left')
```

```python
fulldata
```

```python
pd.crosstab(fulldata["group"], fulldata["landing_page"])
```

```python
correcteddata =fulldata[((fulldata["group"]=="control") & (fulldata["landing_page"]=="old_page")) |((fulldata["group"]=="treatment") & (fulldata["landing_page"]=="new_page"))]
```

```python
correcteddata
```

```python
control_all = correcteddata[(correcteddata['group']=='control')]
```

```python
print(len(control_all))
```

```python
control_cov = correcteddata[((correcteddata['group']=='control') & (correcteddata['converted']==1))]
```

```python
print(len(control_cov))
```

```python
CTR_control = len(control_cov)/len(control_all)
print(CTR_control)
```

```python
treatment_all = correcteddata[(correcteddata['group']=='treatment')]
treatment_cov = correcteddata[((correcteddata['group']=='treatment') & (correcteddata['converted']==1))]
CTR_treatment = len(treatment_cov)/len(treatment_all)
print(CTR_treatment)
```
